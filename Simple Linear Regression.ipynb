Let us explore a simple linear regression in Python using Scikit-learn

```python
# Importing standard packages
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import warnings
warnings.filterwarnings('ignore')
```

```python
# Creating a dummpy dataset of 10 points lying on the straight line: y=2x+1

x = np.arange(10)
y = 2 * x + 1

# The input data for sklearn is 2D: (samples == 10 x features == 1), so formatting the dataset as expected
# by sklearn model.fit()
X = x[:, np.newaxis]

plt.plot(x,y,'X');
```

```python
# Import the lineat regression library
from sklearn.linear_model import LinearRegression

# Instantiating a model object named 'model'
model = LinearRegression()

# Obtain model information to verify
print(model)
```

```python
# Fit the model on the data 
model.fit(X, y)

# Underscore at the end of a variable indicates a fit parameter
print('The coefficient of Linear Regression Line is: ' + str(model.coef_))
print('The intercept of Linear Regression Line is: ' + str(model.intercept_))
```

```python
# Predict the dependent variable using the Linear Regression model
y_pred = model.predict(X)

# Plot the obtained regression model and independent variable datapoints
plt.plot(y_pred, color="red", linewidth=1)
plt.plot(x, y, 'X')
```
